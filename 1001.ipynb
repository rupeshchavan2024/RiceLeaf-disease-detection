```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
import os
import cv2
import random
from skimage import io

# Set paths to your datasets
bacterial_leaf_blight_path ="C:\\Users\\rupes\\PRCP-1001-RiceLeaf\\Data\\Bacterial leaf blight-20200814T055237Z-001\\Bacterial leaf blight" 
brown_spot_path = "C:\\Users\\rupes\\PRCP-1001-RiceLeaf\Data\\Bacterial leaf blight-20200814T055237Z-001\\Bacterial leaf blight"
leaf_smut_path = "C:\\Users\\rupes\\PRCP-1001-RiceLeaf\\Data\\Bacterial leaf blight-20200814T055237Z-001\\Bacterial leaf blight"
import numpy as np

def load_images(directory_path, label):
  
    for file in os.listdir(directory_path):
        img_path = os.path.join(directory_path, file)
        img = cv2.imread(img_path)

        # Resize the image to a consistent size (e.g., 64x64)
        img = cv2.resize(img, (128, 128))

        # Normalize pixel values to be between 0 and 1
        img = img / 255.0

        # Add any additional preprocessing steps here

        images.append(img)
        labels.append(label)

    return images, labels

images = []
labels = []

# Load images and labels for each class
load_images(bacterial_leaf_blight_path, 0)  # Assign label 0 for Bacterial leaf blight
load_images(brown_spot_path, 1)  # Assign label 1 for Brown spot
load_images(leaf_smut_path, 2)  # Assign label 2 for Leaf smut


```

```python
combined_dataset = list(zip(images, labels))
random.shuffle(combined_dataset)
images[:], labels[:] = zip(*combined_dataset)
```

```python
import cv2

# Assuming you have a list of images and labels (images, labels)

# Resize images to a common size (e.g., 64x64 pixels)
resized_images = [cv2.resize(img, (128, 128)) for img in images]

# Convert the list of resized images to a NumPy array
images_np = np.array(resized_images)

# Now you can access the shape attribute
print(images_np.shape)
```

```python
# Convert the list of labels to a NumPy array
labels_np = np.array(labels)

# Now you can access the shape attribute
print(labels_np.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Assuming you have images_np and labels_np from the previous steps

# Flatten images if needed (required for Logistic Regression)
num_samples, height, width, channels = images_np.shape
images_flattened = images_np.reshape(num_samples, -1)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(images_flattened, labels_np, test_size=0.2, random_state=5)

# Create a Logistic Regression model
model = LogisticRegression()

# Train the model
model.fit(X_train, y_train)

# Make predictions on the test set
predictions = model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, predictions)
print("Accuracy:", accuracy)
```

```python
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import StandardScaler
from tensorflow import keras
from tensorflow.keras import layers

# Assuming you have X_train, X_test, y_train, y_test from the train-test split

# Traditional Machine Learning Models
models = {
    'SVM': SVC(),
    'Random Forest': RandomForestClassifier()
}

for model_name, model in models.items():
    # Training
    model.fit(X_train, y_train)
    
    # Prediction
    y_pred = model.predict(X_test)
    
    # Evaluation
    accuracy = accuracy_score(y_test, y_pred)
    print(f"\n{model_name}:\nAccuracy: {accuracy:.4f}")
    print(classification_report(y_test, y_pred))

# Deep Learning Model
# Assuming X_train, X_test are normalized to [0, 1]
# If not, you may need to normalize them using StandardScaler or other techniques

# Define a simple neural network architecture
model = keras.Sequential([
    layers.Flatten(input_shape=(X_train.shape[1:])),  # Assumes X_train is already flattened
    layers.Dense(128, activation='relu'),
    layers.Dense(3, activation='softmax')  # Assuming 3 classes, adjust as needed
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Training
model.fit(X_train, y_train, epochs=20, batch_size=32, validation_split=0.1)

# Evaluation
_, accuracy = model.evaluate(X_test, y_test)
print("\nNeural Network:")
print(f"Accuracy: {accuracy:.4f}")
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from tensorflow import keras
from tensorflow.keras import layers

# Assuming you have X_train, X_test, y_train, y_test from the train-test split

# Reshape flattened input to (64, 64, 3)
X_train_reshaped = X_train.reshape(-1, 128, 128, 3)
X_test_reshaped = X_test.reshape(-1, 128, 128, 3)

# Define a simple CNN architecture
model = keras.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(128, 128, 3)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(128, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dense(244, activation='relu'),
    layers.Dense(3, activation='softmax')  # Assuming 3 classes, adjust as needed
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Training
model.fit(X_train_reshaped, y_train, epochs=50, batch_size=32, validation_split=0.1)

# Evaluation
_, accuracy = model.evaluate(X_test_reshaped, y_test)
print("\nConvolutional Neural Network:")
print(f"Accuracy: {accuracy:.4f}")
```
